```python
import numpy as np
np.random.seed(0)

x = np.linspace(-5, 5, 100)

y = x**3 + np.sin(x)

noise = np.random.normal(0, 10, size =x.shape)

y_data = y + noise

import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(x, y_data, color ='red')
plt.xlabel('x', fontsize =15)
plt.ylabel('y', fontsize =15)
plt.xticks(fontsize =15)
plt.yticks(fontsize =15)
plt.grid()
plt.show()
```

### Linerar Regression Using Matrix Formulation

```python
A = np.hstack([np.ones((len(x), 1)), x.reshape(len(x),1), (x**2).reshape(len(x), 1), (x**3).reshape(len(x), 1), (np.sin(x)).reshape(len(x), 1)])
np.linalg.inv(A.T @ A) @ A.T @ y_data

ATA_inv = np.linalg.inv(A.T@A)
A_transpose = A.T
x_value = ATA_inv @ A_transpose @ y_data

prediction = A@x_value
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(x, y_data, color ='red', label='data')
plt.plot(x, prediction, color ='blue', label='prediction')
plt.xlabel('x', fontsize =15)
plt.ylabel('y', fontsize =15)
plt.xticks(fontsize =15)
plt.yticks(fontsize =15)
plt.grid()
plt.show()
```
