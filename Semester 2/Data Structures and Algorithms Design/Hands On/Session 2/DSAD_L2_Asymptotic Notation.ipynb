# 3n+7

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
fig = plt.figure(figsize=(20,20))
ax1=fig.add_subplot(221)
ax2=fig.add_subplot(222)

x=np.arange(start=1,stop=15,step=1)
xi=x
y=np.arange(start=1,stop=100,step=1)
yi=y
#ax1.set_xticks(xi)
#ax1.set_yticks(yi)

u=(3*x)+7
v=4*x
ax1.set_title('c=4')
ax1.plot(x,u,'r',label='y=3x+7')
for xy in zip(x,u):                                       
    ax1.annotate('(%s, %s)' % xy, xy=xy, textcoords='data',va='top') 

ax1.plot(x,x,'b',label='y=x') 
ax1.plot(x,v,'g',label='y=4x')
idx = np.argwhere((u-v)==0).flatten()
ax1.plot(x[idx], v[idx], 'rD')
ax1.text(x[idx], v[idx],'Intersection Point')
ax1.legend()
ax1.grid(True)

ax2.set_title('c=10')

u=(3*x)+7
v=10*x

ax2.plot(x,u,'r',label='y=3x+7')
for xy in zip(x,u):                                       
    ax2.annotate('(%s, %s)' % xy, xy=xy, textcoords='data',va='top') 

ax2.plot(x,x,'b',label='y=x') 
ax2.plot(x,v,'g',label='y=10x')
idx = np.argwhere((u-v)==0).flatten()
ax2.plot(x[idx], v[idx], 'rD')
ax2.text(x[idx], v[idx],'Intersection Point')
ax2.legend()
ax2.grid(True)
```

# 3x^3+20x^2+5

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
ax.set_title('c=28')

x=np.arange(start=1,stop=5,step=1)
xi=x
y=np.arange(start=1,stop=125,step=1)
yi=y**3
#plt.xticks(xi,xi)
#plt.yticks(yi,yi)

u=(3*(x**3)+(20*(x**2))+5)

v=(x**3)

plt.plot(x,u,'r',label='y=3n^3+20n^2+5')
for xy in zip(x,u):                                       
    ax.annotate('(%s, %s)' % xy, xy=xy, textcoords='data',va='top') 

plt.plot(x,v,'b',label='y=n^3')
for xy in zip(x,v):                                       
    ax.annotate('(%s, %s)' % xy, xy=xy, textcoords='data',va='top')
plt.plot(x,28*(x**3),'g',label='y=28n^3')
for xy in zip(x,28*(x**3)):                                       
    ax.annotate('(%s, %s)' % xy, xy=xy, textcoords='data',va='top')
#idx = np.argwhere((u-v)==0).flatten()
#plt.plot(x[idx], v[idx], 'rD')
#plt.text(x[idx], v[idx],'Intersection Point')
plt.legend()
plt.grid(True)
```
